```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = control.tf([1, 0], [1])
t = np.linspace(0, 200, 10000)
w0 = 1
u = 2*np.cos(w0*t)
G = (s**2 + 1)/(s**2 + 2*s + 4)
y_ss = abs(G.horner(1j*w0))*2*np.cos(w0*t + np.angle(G.horner(1j*w0)))
t, y, x = control.forced_response(G, T=t, U=u)   # lsim in matlab
plt.plot(t, y, label='y')
plt.plot(t, y_ss.T, label='yss')
plt.legend()
```

```python
control.bode((s-1));
```
