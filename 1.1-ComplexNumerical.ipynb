```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


def comet(x, y, duration=2, fps=10, *args, **kwargs):
    fig, ax = plt.subplots()
    ln, = ax.plot([], [], 'r.-', alpha=0.5, *args, **kwargs)
    dot, = ax.plot([], [], marker='o', markersize=10)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

    def init():
        ax.set_xlim(np.min(x)*1.01, np.max(x)*1.01)
        ax.set_ylim(np.min(y)*1.01, np.max(y)*1.01)
        return ln,

    def update(frame):
        i = int(frame*len(x)/(duration*fps))
        ln.set_data(x[:(i + 1)], y[:(i + 1)])
        dot.set_data([x[i]], [y[i]])
        return ln, dot,
    
    plt.close()
    return FuncAnimation(fig, update, frames=range(duration*fps), init_func=init, blit=True, interval=1000/fps)
```

```python
t = np.linspace(0, 2*np.pi, 200)
i = 1j
r = t
z = r*np.exp(i*t)
comet(np.real(z), np.imag(z))
```

```python
i = 1j
th = np.linspace(0, 2*2*np.pi)
z = 10*np.exp(i*th)
comet(th, th, duration=2)
```
